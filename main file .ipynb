```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Customer_Churn.csv')
df.head()
```

# replaceing blanks with 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ",'0')
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"

df['SemiorCitizen'] = df["SeniorCitizen"].apply(conv)
```

# converted 0 and 1 value of senoir citizen to yes/no to make it easier to understand

```python
ax = sns.countplot(x = 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customer by Churn")
plt.show
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customeres", fontsize = 10)
plt.show()
```

# from the given pie chart we can conclude that 26.54% of our customars have out.
# not let's explore the reason behind it

```python
plt.figure(figsize = (4,4))
sns.countplot(x = "gender", data = df,hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
df_percentage = df.groupby("SeniorCitizen")["Churn"].value_counts(normalize=True).mul(100).unstack()

# Create stacked bar chart
df_percentage.plot(kind="bar", stacked=True, figsize=(4,4), colormap="viridis")

# Add labels
for i, bars in enumerate(df_percentage.iterrows()):
    total = sum(bars[1])
    for j, val in enumerate(bars[1]):
        plt.text(i, sum(bars[1][:j+1]) - val/2, f"{val:.1f}%", ha="center", color="white", fontsize=10)

plt.title("Churn by Senior Citizen (Percentage)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
```

# comparetive a greater percentage of people in senior citizen category have churned

```python
# Assuming your DataFrame is called df
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the figure with subplots
plt.figure(figsize=(12, 15))
for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)  # 3 rows x 3 columns
    sns.countplot(data=df, x=col,hue=col, palette="Set2", order=df[col].value_counts().index, legend=False)
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### The majority of customars who do not churn tend to have service like Phoneservice, Internetservice(particularly DSL),and OnlineSecurity enabled OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.
